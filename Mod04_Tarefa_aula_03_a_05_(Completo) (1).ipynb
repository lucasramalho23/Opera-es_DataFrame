Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np
# Presado tutor, você diminuiu muito meu trabalho, porém queria ter a experiencia e a dificuldade de ter gerado essas
# linhas de código sozinho.

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
# criar variável com nome semelhante e depois dividir a var criada antes pelo numero de questões
pc_mat = df_mat /24
pc_por = df_por /18
pc_geo = df_geo /8
pc_ing = df_ing /8
pc_his = df_his /8
pc_fis = df_fis /12
pc_qui = df_qui /12

````

````python
# 2) Seu código aqui
# soma todas as var criadas antes
totalAcertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
totalAcertos
````

````python
# 3) Seu código aqui
# divide o total pelo total de questões que é 90
perventualNota = (totalAcertos/90)*100
perventualNota
````

````python
# 4) Seu código aqui
# Criar uma var que receba da coluna qt_questões o total de acertos que supere  45 e conte quantos foram maiores com o .count()
def aprovados(totalAcertos):
    totalAcertos['aprovados'] = totalAcertos['Qt_acertos']>45
    return totalAcertos[totalAcertos['aprovados']]
aprovados_df = aprovados(totalAcertos)

aprovados(totalAcertos).count()
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# filtro para pegar os menores de 18 anos
df_menor = tabela_vacina[tabela_vacina['paciente_idade']<18].copy()
df_menor.head(4)
# agrupando os estabelecimentos df_menor.groupby('paciente_endereco_nmMunicipio') 
# e somando a quantidade de vacinas ['vacina_descricao_dose'].count().reset_index()
df_menor_municipio = df_menor.groupby(
    'paciente_endereco_nmMunicipio')[
        'vacina_descricao_dose'].count().reset_index()
#trocar o nome da coluna de vacinas_dose por quantiade de vacians aplicadas
# e usando o inplace para atualizar o df original
df_menor_municipio.rename(columns={'vacina_descricao_dose': 'QT_vacinas_aplicadas'},inplace=True)
df_menor_municipio.head(5)

````

````python
# 2)
lista_colunas = df_menor.columns
lista_colunas
````

````python
# 3)
contagem_vacina_nome = df_menor['vacina_categoria_nome'].count()
contagem_vacina_nome
````

````python
# 4)
contagem_estabelecimento_nome = df_menor[['estabelecimento_valor',
'estabelecimento_razaoSocial',
'estalecimento_noFantasia',
'estabelecimento_municipio_codigo',
'estabelecimento_municipio_nome',
'estabelecimento_uf']]

contagem_estabelecimento_nome.columns
````

````python
# 5)
contagem_estabelecimento_nome_loc = df_menor.loc[:,['estabelecimento_valor',
'estabelecimento_razaoSocial',
'estalecimento_noFantasia',
'estabelecimento_municipio_codigo',
'estabelecimento_municipio_nome',
'estabelecimento_uf']]
contagem_estabelecimento_nome_loc.head()
````

````python
#6 
# removendo a linhas com duplicatas e usando o subset para declarar quais eu quero que verifique.
df_lista_dropDuplicates = contagem_vacina_nome_loc.drop_duplicates(subset=
['estabelecimento_valor',
'estabelecimento_razaoSocial',
'estalecimento_noFantasia',
'estabelecimento_municipio_codigo',
'estabelecimento_municipio_nome',
 'estabelecimento_uf'])

#colocando o resultado em uma lista
lista_estabelecimentos_unicos = df_lista_dropDuplicates.values.tolist()

# vamos exibir as informações dessa lista
for estabelecimento in lista_estabelecimentos_unicos:
    print(estabelecimento)
````
